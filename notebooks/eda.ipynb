```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier

df = pd.read_csv('/workspaces/Health-Indicators/notebooks/diabetes_012_health_indicators_BRFSS2015.csv')

# Split the data from 0 to 129,999th row
df_first_half = df.iloc[:130000, :]

# Assuming df_first_half is already defined and contains the first 130,000 rows of your dataset
class_distribution = df_first_half['Diabetes_012'].value_counts(normalize=True)

plt.figure(figsize=(10, 5))
class_distribution.plot(kind='bar')
plt.title('Class Distribution of Diabetes Status')
plt.xlabel('Diabetes Status')
plt.ylabel('Proportion of Dataset')
plt.xticks(ticks=[0, 1, 2], labels=['No Diabetes', 'Prediabetes', 'Diabetes'], rotation=0)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Features')
plt.show()
```

```python
from imblearn.over_sampling import SMOTE

# Apply SMOTE to the target
smote = SMOTE(random_state=42)
X = df_first_half.drop('Diabetes_012', axis=1)
y = df_first_half['Diabetes_012']
X_sm, y_sm = smote.fit_resample(X, y)

# Confirm the class distribution after SMOTE
print(y_sm.value_counts(normalize=True))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.2, random_state=42, stratify=y_sm)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the RandomForestClassifier
rf = RandomForestClassifier(random_state=42)

# Train the classifier on the training set
rf.fit(X_train_scaled, y_train)

# Get feature importances
feature_importances = rf.feature_importances_

print(feature_importances)
```

```python
# Assuming feature_importances is the array you've obtained from the RandomForestClassifier
import pandas as pd

# Convert to Series for easy manipulation
feature_importances_series = pd.Series(feature_importances, index=X_train.columns)

# Convert to percentages
feature_importances_percentages = feature_importances_series * 100

# Sort and display
print(feature_importances_percentages.sort_values(ascending=False))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize and train the RandomForestClassifier on the full scaled dataset
rf_full = RandomForestClassifier(random_state=42)
rf_full.fit(X_train_scaled_full, y_train_full)

# Get feature importances for the full dataset
feature_importances_full = rf_full.feature_importances_
print(feature_importances_full)
```
